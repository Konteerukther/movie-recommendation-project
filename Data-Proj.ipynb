## เช็ค Inconsistent

### จาก unique ที่ไม่เท่ากันระหว่าง movieId กับ title

```python
movie.describe(include='all')
```

```python
#หา title ที่มี movieId ไม่เท่ากับ 1
not_oneId_title = movie["title"].value_counts()[movie["title"].value_counts() != 1]
not_oneId_title
```

```python
not_oneId_title_list = list(not_oneId_title.index)
not_oneId_title_list
```

```python
#แสดง row ของ title นั้น ๆ
not_oneId_movie = movie.loc[movie["title"].isin(not_oneId_title_list)].sort_values(by='title')
not_oneId_movie
```

```python
not_oneId_Id_list = not_oneId_movie["movieId"].unique()
not_oneId_Id_list
```

```python
not_oneId_link = link.loc[link["movieId"].isin(not_oneId_Id_list)]
not_oneId_link
```

```python
#เช็คว่ามีการ rating และ tag หนังที่มีปัญหามั้ย
not_oneId_rating = rating.loc[rating["movieId"].isin(not_oneId_Id_list)]
not_oneId_rating.sort_values(by='movieId')
```

```python
#เช็คว่า Id ที่หาตรงกันมั้ย
set(not_oneId_rating.movieId.unique()) == set(not_oneId_Id_list)
```

```python
not_oneId_tag = tag.loc[tag["movieId"].isin(not_oneId_Id_list)]
not_oneId_tag
```

### มี rating เยอะต้องจัดการ

```python
left = not_oneId_link.set_index("movieId")
right = not_oneId_movie.set_index("movieId")

left.join(right).sort_values(by='title')
```
